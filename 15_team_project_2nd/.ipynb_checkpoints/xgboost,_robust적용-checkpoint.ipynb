```python
import pandas as pd
import numpy as np
import os
import pycaret
from tqdm import tqdm
```

```python
data = pd.read_csv('2013-2018년_가스공급량_기온_습도_기압01.csv')
data
```

```python
from pycaret.regression import *
```

```python
help(setup)
```

```python
sub_2019 = pd.read_csv('./data/2019년_예측_기온_기압_습도.csv')
```

```python
sub_2019
```

```python
models()
```

```python
model_xg = create_model('xgboost')
```

```python
sub_all = pd.DataFrame()
```

```python
data
```

```python
for i in range(7):
    target_data = data[data['구분'] == i]
    target_data = target_data[['month', 'weekday', '시간', '기온', '습도', '기압', '공급량']]
    py_setup = setup(data=target_data, target = '공급량', normalize = True, normalize_method='robust', use_gpu=True, silent=True)

    model_xg = create_model('xgboost')
    model_lgbm = create_model('lightgbm')
    model_cat = create_model('catboost')

    model_xg = tune_model(model_xg)
    model_lgbm = tune_model(model_lgbm)
    model_cat = tune_model(model_cat)

    model_xg = finalize_model(model_xg)
    model_lgbm = finalize_model(model_lgbm)
    model_cat = finalize_model(model_cat)

    prediction_xg = predict_model(model_xg, data = sub_2019[sub_2019['구분'] == i])
    prediction_lgbm = predict_model(model_lgbm, data = sub_2019[sub_2019['구분'] == i])
    prediction_cat = predict_model(model_cat, data = sub_2019[sub_2019['구분'] == i])

    prediction_xg = prediction_xg.reset_index(drop=True)
    prediction_lgbm = prediction_lgbm.reset_index(drop=True)
    prediction_cat = prediction_cat.reset_index(drop=True)

    xg_col = 'prediction_xg_' + str(i)
    lgbm_col = 'prediction_lgbm_' + str(i)
    cat_col = 'prediction_cat_' + str(i)

    sub_all[xg_col] = prediction_xg.Label
    sub_all[lgbm_col] = prediction_lgbm.Label
    sub_all[cat_col] = prediction_cat.Label
```

```python
sub_all
```

```python
sub_average = pd.DataFrame()
```

```python
for i in range(7):
    xg_col = 'prediction_xg_' + str(i)
    lgbm_col = 'prediction_lgbm_' + str(i)
    cat_col = 'prediction_cat_' + str(i)

    average_col = 'average_' + str(i)

    sub_average[average_col] = (sub_all[xg_col] + sub_all[lgbm_col] + sub_all[cat_col]) / 3
```

```python
sub_average
```

```python
sub_2019
```

```python
sub_0 = sub_2019[sub_2019['구분'] == 0]
sub_1 = sub_2019[sub_2019['구분'] == 1].reset_index(drop=True)
sub_2 = sub_2019[sub_2019['구분'] == 2].reset_index(drop=True)
sub_3 = sub_2019[sub_2019['구분'] == 3].reset_index(drop=True)
sub_4 = sub_2019[sub_2019['구분'] == 4].reset_index(drop=True)
sub_5 = sub_2019[sub_2019['구분'] == 5].reset_index(drop=True)
sub_6 = sub_2019[sub_2019['구분'] == 6].reset_index(drop=True)
```

```python
sub_0['공급량'] = sub_average['average_0']
sub_1['공급량'] = sub_average['average_1']
sub_2['공급량'] = sub_average['average_2']
sub_3['공급량'] = sub_average['average_3']
sub_4['공급량'] = sub_average['average_4']
sub_5['공급량'] = sub_average['average_5']
sub_6['공급량'] = sub_average['average_6']
```
