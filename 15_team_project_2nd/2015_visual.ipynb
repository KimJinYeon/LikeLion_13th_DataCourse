```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=total['요일'], y=total['시간(1시)'],
                    mode='markers',
                    name='요일별 난방량'))
fig.update_layout(title='2015년 요일별 난방량',
                   xaxis_title='요일',
                   yaxis_title='난방')
fig.show()
```
