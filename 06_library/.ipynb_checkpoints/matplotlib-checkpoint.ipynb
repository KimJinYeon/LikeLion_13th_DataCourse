```python
# 요일별 식사금액은 얼마나 될까?
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

* 식사금액은 토요일과 일요일이 많다
* 현재 데이터는 목, 금, 토, 일의 데이터가 있다.
* 토요일과 일요일을 비교해 본다면 상대적으로 일요일의 지출 식사 금액이 더 많다.

# 요일별 팁

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```

* 최소 팁은 2원인 것 같다.
* 일요일에 팁을 더 많이 준다.
* 토요일에는 팁의 편차가 크다.

# 요일별 식사인원

```python
sns.boxplot(x='day', y='size', data=tips)
plt.show()
```

* 토요일보다 일요일에 더 많은 인원으로 식사하는 경향이 있다.
* 목요일과 금요일은 편차가 크다.
