```python
import pandas as pd
from datetime import date, timedelta
```

```python
corona = pd.read_csv('./data/coronaboard(2021-09-28).csv')
vaccine = pd.read_csv('./data/bloomberg(2021-09-28).csv')
country = pd.read_csv('./data/country.csv')
```

```python
jandam = corona.loc[corona['국가'] == '잔담🚢', :]
jandam
```

```python
corona.drop(jandam.index, inplace=True)
```

```python
corona['국가']
```

```python
corona['국가코드'] = corona['국가'].str[-2:]
```

```python
corona['국가'] = corona['국가'].str[:-2]
```

```python
corona.head()
```

```python
corona.columns
```

```python
col = ['국가', '국가코드', '확진자', '치료중', '위중증', '사망자', '완치', '치명(%)', '완치(%)', '발생률', '인구수']
```
