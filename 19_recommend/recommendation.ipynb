```python
columns = ['item_id', 'movie title', 'release date', 'video release date', 'IMDb URL', 
           'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 
           'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 
           'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 
           'Sci-Fi', 'Thriller', 'War', 'Western']

movies = pd.read_csv('./data/ml-100k/u.item', sep='|', names=columns, encoding='latin-1')
```

```python
movies.head(100)
```

```python
movies.info()
```

```python
print(len(df.item_id.unique()))
print(len(movies.item_id.unique()))
```

```python
movie_names = movies[['item_id', 'movie title']]
```

```python
c_movies_data = pd.merge(df, movie_names, on='item_id')
```

```python
c_movies_data.head(100)
```

### 영화 하나 선택, 관련 유사한 영화 10편 추천

```python
rating_c = c_movies_data.pivot_table(values="rating",
                                   index="user_id",
                                   columns="movie title",
                                   fill_value=0)
```

```python
rating_c.head()
```

```python
X = rating_c.T
X.head()
```

### 차원 축소

```python
SVD = TruncatedSVD(n_components=12, random_state=5)
resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape
```

```python
resultant_matrix
```

```python
corr_mat = np.corrcoef(resultant_matrix)
print( corr_mat.shape )
corr_mat
```

```python
rating_c.columns.get_loc("Star Wars (1977)")
```

```python
col_idx = rating_c.columns.get_loc("Star Wars (1977)")
col_idx
```

```python
type(col_idx)
```

```python
corr_specific = corr_mat[col_idx]
corr_specific
```

```python
result = pd.DataFrame({'corr_specific':corr_specific, 'Movies': rating_c.columns})
print(result.shape)
result.head()
```

```python
result.sort_values('corr_specific', ascending=False).head(11)
```

```python
col_idx = rating_c.columns.get_loc("101 Dalmatians (1996)")
corr_specific = corr_mat[col_idx]
result = pd.DataFrame({'corr_specific':corr_specific, 'Movies': rating_c.columns})
result.sort_values('corr_specific', ascending=False).head(16)
```

```python
def getRecommendation(movie_name, number):
    col_idx = rating_c.columns.get_loc(movie_name)
    corr_specific = corr_mat[col_idx]
    result = pd.DataFrame({'corr_specific':corr_specific, 'Movies': rating_c.columns})
    return result.sort_values('corr_specific', ascending=False).head(number)
```

```python
target = getRecommendation('Silence of the Lambs, The (1991)', 11)
target
```
