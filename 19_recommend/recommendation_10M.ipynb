```python
import pandas as pd
import numpy as np
import sklearn
```

```python
columns = ['item_id', 'movie_name', 'genre']
movie = pd.read_csv('./data/ml-10M100K/movies.dat', sep='::', names=columns)
movie = movie.drop(columns='genre').copy()
movie
```

```python
columns = ['user_id', 'item_id', 'rating', 'timestamp']
rating = pd.read_csv('./data/ml-10M100K/ratings.dat', sep='::', names=columns)
rating
```

```python
print(len(rating.item_id.unique()))
print(len(movie.item_id.unique()))
```

```python
movies_data = pd.merge(rating, movie, on='item_id')
```
