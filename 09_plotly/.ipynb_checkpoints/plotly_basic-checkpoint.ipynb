```python
import plotly
import cufflinks as cf
import pandas as pd
import numpy as np
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
print(plotly.__version__)
print(cf.__version__)
```

```python
dat = np.random.rand(100,4)
df = pd.DataFrame(dat, columns = 'A B C D'.split())
df.head()
```

```python
df.loc[:, 'A':'C'].iplot(kind = 'line',
             xTitle='데이터 idx',
             yTitle='y축 - 값',
             title='100개의 값의 선 그래프')
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers+text', size=20)
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers+text', size=20,
        color='green', symbol='diamond')
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers+text', size=20,
        color='blue', symbol='pentagon')
```

```python
exercise = sns.load_dataset('exercise')
```

```python
exercise
```

```python
exercise.time.unique()
```

```python
time_list = []
k=0
for i in exercise['time']:
    i = int(i.replace('min', '').strip())
    time_list.append(i)
```

```python
exercise.loc[:,['time']] = time_list
```

```python
genre_list = []
k=0
for i in exercise['kind']:
    genre_list.append(i)
```

```python
exercise.loc[:,['genre']] = genre_list
```

```python
exercise
```

```python
exercise['time'][0]
```

```python
exercise
```

```python
exercise.iplot(kind='scatter', x='kind', y='pulse', mode='markers')
```
